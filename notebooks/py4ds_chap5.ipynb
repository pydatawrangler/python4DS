```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/byuidatascience/data4python4ds/master/data-raw/flights/flights.csv"
flights = pd.read_csv(url)
```

```python
flights.head()
```

```python
flights.info()
```

```python
flights["time_hour"] = pd.to_datetime(flights["time_hour"], format = "%Y-%m-%dT%H:%M:%SZ")
```

```python
flights["time_hour"]
```

```python
(flights
 .query("dest == 'IAH'")
 .groupby(["year", "month", "day"])
 [["arr_delay"]].mean()
)
```

```python
flights.query("month == 1 and day == 1")
```

```python
flights.sort_values(["year", "month", "day", "dep_time"])
```

```python
(flights
 .query("dep_delay <= 10 and dep_delay >= -10")
 .sort_values("arr_delay", ascending=False)
 .iloc[[0, 1, 2]]
)
```

```python
(flights
 .assign(
     gain = lambda df_: df_["dep_delay"] - df_["arr_delay"],
     speed = lambda df_: df_["distance"] / df_["air_time"] * 60
 )
)
```

```python
flights.select_dtypes("int")
```

```python
(flights
 .groupby("month")
 [["dep_delay"]]
 .mean()
)
```

```python
(flights
 .assign(max_delay_month = lambda df_: df_.groupby("month")["arr_delay"].transform("max"))
 .assign(delay_frac_of_max = lambda df_: df_["arr_delay"] / df_["max_delay_month"])
 [["year", "month", "day", "arr_delay", "max_delay_month", "delay_frac_of_max"]]
).head()
```
